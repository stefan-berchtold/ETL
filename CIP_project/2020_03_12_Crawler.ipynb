```python
import requests
import bs4
import re

r = requests.get("https://de.wikipedia.org/wiki/Liste_von_Fluggesellschaften")

from bs4 import BeautifulSoup


doc = BeautifulSoup(r.text, "html.parser")
doc.find_all("table", class_="wikitable sortable")
My_table = doc.find("table", {"class":"wikitable sortable"}) 
td_table = My_table.findAll("td")
iata = td_table[2::7]
rufzeichen = td_table[3::7]

rufzeichen = str(rufzeichen)

splitted_rufzeichen = rufzeichen.split(",")

rufzeichen_splitted_one = [w[:-6] for w in splitted_rufzeichen]

rufzeichen_splitted_two = [w[5::] for w in rufzeichen_splitted_one]

#rufzeichen_splitted_two

iata = str(iata)

splitted_iata = iata.split(",")

iata_splitted_one = [w[:-6] for w in splitted_iata]

iata_splitted_two = [w[5::] for w in iata_splitted_one]

#iata_splitted_two

```

```python
import pandas as pd
df = pd.DataFrame()
df['IATA'] = iata_splitted_two
df['Rufzeichen'] = rufzeichen_splitted_two
df
```
